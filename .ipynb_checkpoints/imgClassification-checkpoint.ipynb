```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(gray, cmap="gray")
```

### Haar Cascade (to detect face)

```python
# keninde göre bir mantığı var mesela göz ile burun arasındaki 
# ton farkı sayesinde bir şekil tespit ediyor gibi...
```

```python
face_cascade = cv2.CascadeClassifier("haar-opencv/haarcascade_frontalface_default.xml")
eye_cascade = cv2.CascadeClassifier("haar-opencv/haarcascade_eye.xml")

faces = face_cascade.detectMultiScale(gray, 1.3, 5)
faces
```

```python
# x y width height
```

```python
(x,y,w,h) = faces[0]
x,y,w,h
```

```python
# w ve h mesafe değeri aslında + diye ondan ekliyoruz 
```

```python
face_img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
plt.imshow(face_img)
```

### Eye Detection (Open-cv nin sayfasında var bu kod)

```python
cv2.destroyAllWindows()
for (x, y, w, h) in faces:
    face_img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,0,0), 2)
    roi_gray = gray[y:y+h, x:x+w]
    roi_color = face_img[y:y+h, x:x+w]
    eyes = eye_cascade.detectMultiScale(roi_gray)
    for (ex, ey, ew, eh) in eyes:
        cv2.rectangle(roi_color, (ex,ey), (ex+ew, ey+eh), (0,255,0), 2)
        
plt.figure()
plt.imshow(face_img, cmap="gray")
plt.show()
```

```python
%matplotlib inline
plt.imshow(roi_color, cmap="gray")
```

```python
def get_cropped_image_if_2_eyes(image_path):
    img = cv2.imread(image_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray, 1.3, 5)
    for (x, y, w, h) in faces:
        roi_gray = gray[y:y+h, x:x+w]
        roi_color = img[y:y+h, x:x+w]
        eyes = eye_cascade.detectMultiScale(roi_gray)
        if len(eyes) >=2:
            return roi_color
```

```python
img = cv2.imread("test-images/Cristiano_Ronaldo_2018.jpg")
plt.imshow(img);
```

```python
cropped_img = get_cropped_image_if_2_eyes("test-images/Cristiano_Ronaldo_2018.jpg")
plt.imshow(cropped_img);
```
